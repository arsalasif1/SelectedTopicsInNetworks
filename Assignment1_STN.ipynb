```python
## defining function to hash files for Syntax to use = sHashFile(sFilePath, iBlockSize, sEncryptionType)

def sHashFile(sFilePath, iBlockSize, sEncryptionType):
  BLOCK_SIZE = iBlockSize # The size of each read from the file

  file_hash = hashlib.new(sEncryptionType)
  with open(sFilePath, 'rb') as f: # Open the file to read it's bytes
      fb = f.read(BLOCK_SIZE) # Read from the file. Take in the amount declared above
      while len(fb) > 0: # While there is still data being read from the file
          file_hash.update(fb) # Update the hash
          fb = f.read(BLOCK_SIZE) # Read the next block from the file
      f.close()
      return file_hash.hexdigest()

############################################################


# creating text file and  adding sentence to it
sFileName = "mytextfile_1.txt"
with open(sFileName, "w") as xFile:
  xFile.write("We do, what we do best, we improvise")
  xFile.close()

sHexOfFile = sHashFile(sFilePath = sFileName, iBlockSize = 1024, sEncryptionType = "sha256")  # Get the hexadecimal digest of the file

print ("Original File : " + sHexOfFile )

# opening text file and changing the sentence
sFileName = "mytextfile_1.txt"
with open(sFileName, "w") as xFile:
  xFile.write("We do, what we do best, we improvise!") # Exclamation mark (!) added at the end to observe avalanche effect
  xFile.close()

sHexOfFile = sHashFile(sFilePath = sFileName, iBlockSize = 1024, sEncryptionType = "sha256")
print ("Changed File : " + sHexOfFile ) # Get the hexadecimal digest of the file
```

```python
import math
def sTwoStringHash(s1, s2, sEncryptionType):
 # print(s1,s2)

  sHash = s1+s2
  return hashlib.new(sEncryptionType, sHash.encode()).hexdigest()


def sMarkleTree(sListOfStrings, sEncryptionType):
  if(math.log2(len(sListOfStrings)).is_integer()): #power of 2 check
    sHash_list=[] #list for processing hash
    sMyList = sListOfStrings #list for input
    #loop for hashing first layer
    for i in range(len(sMyList)):
      # hash_arr.append(arr[i]+"h")
      sHash_list.append(hashlib.new(sEncryptionType, sMyList[i].encode()).hexdigest())
    print(sHash_list)
    #loop for hashing consequent layeres until root is achived
    while(len(sHash_list)!=1):
      sMyList = sHash_list
      length = len(sHash_list)
      # loop for getting hashes of pairs eg. H(12) H(34)...
      for i in range(int(len(sHash_list)/2)):
        sHash_list.append(sTwoStringHash(sMyList[i*2], sMyList[2*i+1], sEncryptionType))
      # loop for deleting previous pairs...
      for i in range(length):
        sHash_list.pop(0)
      # printing each layer
      print(sHash_list)
    return(sHash_list[0])
  else:
    print("Cannot build merkle tree the string should be in pow 2, try changing input by copying the last inputs")
    return(0)




  #print(len(ListOfStrings))

sRandomString = ["apple", "banana","apple", "banana","apple", "banana","apple", "banana"]
print("Merkle Tree: ")
print("Root: " + sMarkleTree(sRandomString, "sha256"))
```
